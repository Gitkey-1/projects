```python
# Make copies of original pieces of data
archive_clean = df_archive.copy()
image_clean = df_image_predictions.copy()
tweet_clean = df_tweet_json.copy()
```

### Tidness Issue 
- #### #1:Each various stage of dogs in "df_archive" table shouled be in one column.

#### Define
- ##### combine four columns to one.

#### Code

```python
# replace all none values to empty space
stages = ['doggo', 'floofer', 'pupper', 'puppo']
for i in stages:
    archive_clean[i] = archive_clean[i].replace('None', '')
```

```python
# combine all stages in one column
archive_clean['stages'] = archive_clean['doggo'] + archive_clean['floofer'] + archive_clean['pupper'] + archive_clean['puppo'] 

# drop 'doggo', 'floofer', 'pupper', 'puppo' columns
archive_clean = archive_clean.drop(['doggo','floofer','pupper','puppo'], axis = 1)

# use np.nan to fill the empty spots
archive_clean['stages'] = archive_clean['stages'].replace('', np.nan)
```

### Insights:  #1 Top 10 rates dogs.

```python
# add rates column to DataFrame.
rates = df_2['rating_numerator'] / df_2['rating_denominator']

df_2['rates'] = rates

pd.set_option('max_columns', None)

top_10 = df_2.sort_values(by=['rates'],ascending=False).head(10)
top_10
```

### Visualization

```python
# image of top rated dog. 
Image(url = 'https://pbs.twimg.com/media/CmgBZ7kWcAAlzFD.jpg')
```

```python
#plot of top 10 dogs
x = top_10['name']
y = top_10['rates']

plt.bar(x,y)
plt.xlabel('names')
plt.ylabel('rates')
plt.title('Top 10 Rated Dogs')
plt.show();
```

>The dog named Atticus achieve rate of 177.6, is the top rated dog.

### Insights:  #2 Monthly dogs rates amount.

```python
# Group timestamp by month.
time_m =  df_2.groupby(pd.Grouper(key = 'timestamp', freq = "M")).count().reset_index()
time_m = time_m[['timestamp', 'tweet_id']]
time_m
```

### Visualization

```python
# line chart for dog rate amount.
plt.figure(figsize=(8, 5));
plt.xlim([datetime.date(2015, 11, 30), datetime.date(2017, 7, 30)])
plt.plot(time_m.timestamp, time_m.tweet_id)
plt.title('Monthly dogs rates amount')
plt.xlabel('Month')
plt.ylabel('Monthly Total Count');
```

 > From the line chart above, we can see the trend of amounts is dropping from first few month, down to around 50 times of rate.
